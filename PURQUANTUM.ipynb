```python
import os, warnings, random
warnings.filterwarnings("ignore")
os.environ["TOKENIZERS_PARALLELISM"] = "true"

import numpy as np
import pandas as pd

SEED = 0
random.seed(SEED)
np.random.seed(SEED)

BATCH_SIZE = 8
EPOCHS = 30
SHOTS = 2048

# control "mărimea circuitului"
MAX_WORDS_1 = 12
MAX_WORDS_2 = 12
```

```python
CSV_PATH = "/content/redudancy.csv"  # schimbă dacă e alt nume

df = pd.read_csv(CSV_PATH)
df = df[["sentence1", "sentence2", "label"]].dropna()
df["label"] = df["label"].astype(int)
df.head()
```

```python
def train_val_split(df, val_ratio=0.2, seed=SEED):
    idx = np.arange(len(df))
    rng = np.random.default_rng(seed)
    rng.shuffle(idx)
    cut = int(len(idx) * (1 - val_ratio))
    train_idx, val_idx = idx[:cut], idx[cut:]
    return df.iloc[train_idx].reset_index(drop=True), df.iloc[val_idx].reset_index(drop=True)

train_df, val_df = train_val_split(df, val_ratio=0.2)
len(train_df), len(val_df)
```

```python
def truncate_words(s: str, max_words: int) -> str:
    toks = str(s).strip().split()
    return " ".join(toks[:max_words])

def one_hot_2(label01: int):
    # ca în tutorial: [t, 1-t]
    t = int(label01)
    return [t, 1 - t]
```

```python
from lambeq import BobcatParser, AtomicType, UnifyCodomainRewriter

parser = BobcatParser(verbose="text")  # poți pune "suppress" ca să fie mai silențios
s_type = AtomicType.SENTENCE

unify = UnifyCodomainRewriter(output_type=s_type)

def make_pair_diagrams(df_in: pd.DataFrame):
    # 1) truncăm separat
    s1 = [truncate_words(x, MAX_WORDS_1) for x in df_in["sentence1"].tolist()]
    s2 = [truncate_words(x, MAX_WORDS_2) for x in df_in["sentence2"].tolist()]
    labels = [one_hot_2(y) for y in df_in["label"].tolist()]

    # 2) parse în batch (mai rapid): flatten + reparing
    flat = s1 + s2
    raw = parser.sentences2diagrams(flat, suppress_exceptions=True)

    d1 = raw[:len(s1)]
    d2 = raw[len(s1):]

    pair_diags = []
    pair_labels = []

    for a, b, lab in zip(d1, d2, labels):
        if a is None or b is None:
            continue
        # normal_form pe fiecare + combinare
        pair = (a.normal_form() @ b.normal_form())
        # unește cele 2 ieșiri într-una singură (altfel ai 2 fire libere)
        pair = unify(pair).normal_form()
        pair_diags.append(pair)
        pair_labels.append(lab)

    return pair_diags, pair_labels

train_diags, train_labels = make_pair_diagrams(train_df)
val_diags, val_labels = make_pair_diagrams(val_df)

len(train_diags), len(val_diags)
```

```python
from lambeq import IQPAnsatz, RemoveCupsRewriter

remove_cups = RemoveCupsRewriter()

# mapare tipuri -> număr qubiți
# (îți păstrează și NOUN și SENTENCE ca 1 qubit; poți reduce dacă e prea mare)
ansatz = IQPAnsatz(
    {AtomicType.NOUN: 1, AtomicType.SENTENCE: 1, AtomicType.PREPOSITIONAL_PHRASE:0},
    n_layers=1,
    n_single_qubit_params=3
)

train_circuits = [ansatz(remove_cups(d)) for d in train_diags]
val_circuits   = [ansatz(remove_cups(d)) for d in val_diags]

# optional: vezi un circuit
train_circuits[0].draw(figsize=(10, 6))
```

```python
from pytket.extensions.qiskit import AerBackend
from lambeq import TketModel, QuantumTrainer, SPSAOptimizer
from lambeq import BinaryCrossEntropyLoss, Dataset

backend = AerBackend()
backend_config = {
    "backend": backend,
    "compilation": backend.default_compilation_pass(2),
    "shots": SHOTS
}

all_circuits = train_circuits + val_circuits
model = TketModel.from_diagrams(all_circuits, backend_config=backend_config)

bce = BinaryCrossEntropyLoss()
acc = lambda y_hat, y: np.sum(np.round(y_hat) == y) / len(y) / 2  # ca în tutorial

trainer = QuantumTrainer(
    model,
    loss_function=bce,
    epochs=EPOCHS,
    optimizer=SPSAOptimizer,
    optim_hyperparams={"a": 0.05, "c": 0.06, "A": 0.001 * EPOCHS},
    evaluate_functions={"acc": acc},
    evaluate_on_train=True,
    verbose="text",
    log_dir="logs_qnlp_pairs",
    seed=SEED,
)

train_dataset = Dataset(train_circuits, train_labels, batch_size=BATCH_SIZE)
val_dataset   = Dataset(val_circuits, val_labels, shuffle=False)

trainer.fit(
    train_dataset,
    val_dataset,
    early_stopping_criterion="acc",
    early_stopping_interval=5,
    minimize_criterion=False
)
```

```python
model.load(trainer.log_dir + "/best_model.lt")
val_acc = acc(model(val_circuits), val_labels)
print("Validation accuracy:", float(val_acc))
```

```python
# ===========================
# ONE-CELL Colab script (run)
# ===========================
# CSV format: sentence1,sentence2,label (label in {0,1})
# Put your CSV in /content/data.csv (or change CSV_PATH below)


import os, warnings, random
warnings.filterwarnings("ignore")
os.environ["TOKENIZERS_PARALLELISM"] = "true"

import numpy as np
import pandas as pd

from lambeq import (
    BobcatParser, AtomicType,
    UnifyCodomainRewriter, RemoveCupsRewriter,
    IQPAnsatz, TketModel,
    QuantumTrainer, SPSAOptimizer,
    BinaryCrossEntropyLoss, Dataset
)
from pytket.extensions.qiskit import AerBackend

# -----------------------------
# Settings (tune here)
# -----------------------------
CSV_PATH = "/content/redudancy.csv"   # <-- change if needed

VAL_RATIO = 0.2
BATCH_SIZE = 8
EPOCHS = 50
SHOTS = 8192

# keep circuits small:
MAX_WORDS_1 = 12
MAX_WORDS_2 = 12

# stronger than the tutorial defaults:
N_LAYERS = 2
N_SINGLE_Q_PARAMS = 6

# run multiple seeds to reduce SPSA randomness:
SEEDS = [0, 1, 2, 3, 4]

# SPSA hyperparams (often need tuning)
SPSA_A = 0.02
SPSA_C = 0.10
# A is set relative to epochs below

# -----------------------------
# Helpers
# -----------------------------
def truncate_words(s: str, max_words: int) -> str:
    toks = str(s).strip().split()
    return " ".join(toks[:max_words])

def one_hot_2(label01: int):
    t = int(label01)
    return [t, 1 - t]

def train_val_split(df_in, val_ratio=0.2, seed=0):
    idx = np.arange(len(df_in))
    rng = np.random.default_rng(seed)
    rng.shuffle(idx)
    cut = int(len(idx) * (1 - val_ratio))
    return df_in.iloc[idx[:cut]].reset_index(drop=True), df_in.iloc[idx[cut:]].reset_index(drop=True)

def acc_fn(y_hat, y):
    # y_hat, y are arrays shaped (N,2); round per entry like the tutorial
    y_hat = np.array(y_hat)
    y = np.array(y)
    return float(np.sum(np.round(y_hat) == y) / len(y) / 2)

# -----------------------------
# Load CSV
# -----------------------------
df = pd.read_csv(CSV_PATH)
df = df[["sentence1", "sentence2", "label"]].dropna()
df["label"] = df["label"].astype(int)

print("Rows:", len(df))
print("Label distribution:\n", df["label"].value_counts())

# -----------------------------
# Build parser + rewriters + ansatz (shared)
# -----------------------------
parser = BobcatParser(verbose="suppress")
s_type = AtomicType.SENTENCE
unify = UnifyCodomainRewriter(output_type=s_type)
remove_cups = RemoveCupsRewriter()

ansatz = IQPAnsatz(
    {AtomicType.NOUN: 1, AtomicType.SENTENCE: 1, AtomicType.PREPOSITIONAL_PHRASE:0},
    n_layers=N_LAYERS,
    n_single_qubit_params=N_SINGLE_Q_PARAMS
)

backend = AerBackend()
backend_config = {
    "backend": backend,
    "compilation": backend.default_compilation_pass(2),
    "shots": SHOTS
}

def make_pair_diagrams(df_in: pd.DataFrame):
    s1 = [truncate_words(x, MAX_WORDS_1) for x in df_in["sentence1"].tolist()]
    s2 = [truncate_words(x, MAX_WORDS_2) for x in df_in["sentence2"].tolist()]
    labels = [one_hot_2(y) for y in df_in["label"].tolist()]

    flat = s1 + s2
    raw = parser.sentences2diagrams(flat, suppress_exceptions=True)
    d1 = raw[:len(s1)]
    d2 = raw[len(s1):]

    pair_diags, pair_labels = [], []
    dropped = 0
    for a, b, lab in zip(d1, d2, labels):
        if a is None or b is None:
            dropped += 1
            continue
        pair = (a.normal_form() @ b.normal_form())
        pair = unify(pair).normal_form()  # merge two sentence outputs into one
        pair_diags.append(pair)
        pair_labels.append(lab)

    return pair_diags, pair_labels, dropped

def diagrams_to_circuits(diags):
    return [ansatz(remove_cups(d)) for d in diags]

# -----------------------------
# Run multiple seeds
# -----------------------------
results = []
for run_seed in SEEDS:
    print("\n" + "="*60)
    print(f"RUN seed={run_seed}")
    random.seed(run_seed)
    np.random.seed(run_seed)

    train_df, val_df = train_val_split(df, VAL_RATIO, seed=run_seed)

    train_diags, train_labels, drop_tr = make_pair_diagrams(train_df)
    val_diags,   val_labels,   drop_va = make_pair_diagrams(val_df)

    print(f"After parsing: train={len(train_diags)} (dropped {drop_tr}), val={len(val_diags)} (dropped {drop_va})")

    if len(train_diags) < 10 or len(val_diags) < 5:
        print("Too few examples after parsing; try increasing MAX_WORDS or check your sentences language/format.")
        results.append(np.nan)
        continue

    train_circuits = diagrams_to_circuits(train_diags)
    val_circuits   = diagrams_to_circuits(val_diags)

    # One model per run
    all_circuits = train_circuits + val_circuits
    model = TketModel.from_diagrams(all_circuits, backend_config=backend_config)

    trainer = QuantumTrainer(
        model,
        loss_function=BinaryCrossEntropyLoss(),
        epochs=EPOCHS,
        optimizer=SPSAOptimizer,
        optim_hyperparams={
            "a": SPSA_A,
            "c": SPSA_C,
            "A": 0.01 * EPOCHS
        },
        evaluate_functions={"acc": acc_fn},
        evaluate_on_train=True,
        verbose="text",
        log_dir=f"logs_qnlp_pairs_seed_{run_seed}",
        seed=run_seed,
    )

    train_dataset = Dataset(train_circuits, train_labels, batch_size=BATCH_SIZE)
    val_dataset   = Dataset(val_circuits, val_labels, shuffle=False)

    trainer.fit(
        train_dataset,
        val_dataset,
        early_stopping_criterion="acc",
        early_stopping_interval=5,
        minimize_criterion=False
    )

    # Load best and eval
    model.load(trainer.log_dir + "/best_model.lt")
    val_pred = model(val_circuits)
    val_acc = acc_fn(val_pred, val_labels)
    print(f"VAL acc (seed={run_seed}): {val_acc:.4f}")
    results.append(val_acc)

# -----------------------------
# Summary
# -----------------------------
vals = np.array([x for x in results if not np.isnan(x)], dtype=float)
print("\n" + "="*60)
print("DONE")
print("Seeds:", SEEDS)
print("Val accuracies:", results)
if len(vals):
    print(f"Mean val acc: {vals.mean():.4f} | Std: {vals.std(ddof=1) if len(vals)>1 else 0.0:.4f} | Best: {vals.max():.4f}")
else:
    print("No valid runs (likely too many sentences failed to parse).")
```
